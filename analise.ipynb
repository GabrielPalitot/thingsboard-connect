```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("data.csv")  

df['timestamp'] = pd.to_datetime(df['timestamp'])

plt.figure(figsize=(12, 6))
plt.plot(df['timestamp'], df['detector_chuva'], label='Detector de Chuva')
plt.plot(df['timestamp'], df['humidity_dht11'], label='Umidade do Ar')
plt.plot(df['timestamp'], df['temperature_dht11'], label='Temperatura')
plt.plot(df['timestamp'], df['umidade_solo'], label='Umidade do Solo')
plt.xlabel('Tempo')
plt.ylabel('Valor')
plt.title('Sensores ao Longo do Tempo')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='umidade_categoria', palette='coolwarm')
plt.title('Distribuio de Categorias de Umidade do Solo')
plt.xlabel('Categoria')
plt.ylabel('Frequncia')
plt.show()

plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='detector_categoria', palette='coolwarm')
plt.title('Distribuio de Categorias de Detector de Chuva')
plt.xlabel('Categoria')
plt.ylabel('Frequncia')
plt.show()
```
